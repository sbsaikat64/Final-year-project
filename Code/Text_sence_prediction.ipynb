**Loading text sentiment model**

```python
fname = '/content/drive/MyDrive/test/finalized_text_model.sav'
loaded_model_text = pickle.load(open(fname, 'rb'))
tokenname = '/content/drive/MyDrive/test/text_model_token.sav'
text_tokenizer = pickle.load(open(tokenname, 'rb'))
```

**Loading emoji sentiment model**

```python
fname1 = '/content/drive/MyDrive/test/finalized_emoji_model.sav'
loaded_model_emoji = pickle.load(open(fname1, 'rb'))
tokenname1 = '/content/drive/MyDrive/test/emoji_model_token.sav'
emoji_tokenizer = pickle.load(open(tokenname1, 'rb'))
```

**Taking input from user**

```python
text = input()
```

**Formating the input to match models needs**

```python
emo = text[-1]
txt =[]
txt.append(text[:-1])
#print(emoji.demojize(emo))
#print(txt)
emo = emoji.demojize(emo)
emo = re.sub('_',' ',emo)
emo = re.sub('-',' ',emo)
emo = emo[1:-1]
emo = np.array([emo])
```

**Predicting sentiments using model**

```python
#vectorizing the text by the pre-fitted tokenizer instance
txt = text_tokenizer.texts_to_sequences(txt)
#padding the text to have exactly the same shape as `embedding_2` input
txt = pad_sequences(txt, maxlen=28, dtype='int32', value=0)
#print(twt)
text_sentiment = loaded_model_text.predict(txt,batch_size=1)

emo = emoji_tokenizer.texts_to_sequences(emo)
#padding the text to have exactly the same shape as `embedding_2` input
emo = pad_sequences(emo, maxlen=5, dtype='int32', value=0)
#print(twt)
emoji_sentiment = loaded_model_emoji.predict(emo,batch_size=1)
```

**Printing the output**

```python
print("[[INPUT]]")
print(text, end="\n\n")

print("[[meaning of the emoji]]\n"+emoji.demojize(text[-1]))
#print(text_sentiment)

print("\n\n[[Tone of the text part is]]", end = "\n")
if(np.argmax(text_sentiment) == 0): print("negative"); text_sentiment = 'negative'
elif(np.argmax(text_sentiment) == 1): print("neutral"); text_sentiment = 'neutral'
elif (np.argmax(text_sentiment) == 2): print("positive"); text_sentiment = 'positive'

#print(np.argmax(emoji_sentiment))
print("\n\n[[Tone of the emoji is]]", end = "\n")
if(np.argmax(emoji_sentiment) <= 1): print("negative"); emoji_sentiment = "negative"
elif(np.argmax(emoji_sentiment) == 2): print("neutral"); emoji_sentiment = "neutral"
elif (np.argmax(emoji_sentiment) >= 3): print("positive"); emoji_sentiment = "positive"

if(text_sentiment == "negative" and emoji_sentiment == "negative"): print("\n\n[[Overall sentiment of the sentence is]]\nnegative.")
elif(text_sentiment == "positive" and emoji_sentiment == "positive"): print("\n\n[[Overall sentiment of the sentence is]]\npositive.")
elif(text_sentiment == "neutral"):
    if(emoji_sentiment == "positive"): print("\n\n[[Overall sentiment of the sentence is]]\npositive.")
    elif(emoji_sentiment == "negative"): print("\n\n[[Overall sentiment of the sentence is]]\nnegative.")
elif((text_sentiment == "negative" and emoji_sentiment == "positive") or (text_sentiment == "positive" and emoji_sentiment == "negative")):
    print("\n\n[[Overall sentiment of the sentence is]]\nambigious.")
```
